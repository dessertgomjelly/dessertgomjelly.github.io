```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# 취득학점의 합 및 학점점수의 합
total_credits = I.groupby('학년')['취득학점'].sum()
total_grade_points = I.groupby('학년')['학점점수'].sum()

# 학점 평균 계산
grade_point_average = total_grade_points / total_credits


# 그래프 그리기
plt.figure(figsize=(8, 6))

# 그래프 그리기
plt.plot(grade_point_average.index, grade_point_average.values, marker='x', linestyle='-', color='black')

# 기울기에 따라 색상 설정
colors = []

for i in range(1, len(grade_point_average)):
    slope = grade_point_average.iloc[i] - grade_point_average.iloc[i - 1]
    if slope > 0:
        color = 'blue'  # 기울기가 양수일 때 파란색, 기울기의 변화 정도에 따라 투명도 적용
    else:
        color = 'red' # 기울기가 음수일 때 빨간색, 기울기의 변화 정도에 따라 투명도 적용
    colors.append(color)

# 화살표로 기울기 변화 표시
for i in range(1, len(grade_point_average)):
    plt.annotate('', xy=(i, grade_point_average.iloc[i]), xytext=(i - 0.3, grade_point_average.iloc[i - 1]),
                 arrowprops=dict(arrowstyle='->', color=colors[i - 1]))

plt.title('학년 별 학점 평균')
plt.xlabel('학년')
plt.ylabel('학점 평균')
plt.xticks(grade_point_average.index)
plt.grid(axis='y')

plt.show()
```
